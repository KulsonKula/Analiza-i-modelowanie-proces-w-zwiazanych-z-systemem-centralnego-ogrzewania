```python
import pandas as pd
import numpy as np
import datetime
import os 
import glob 
np.set_printoptions(precision=5, suppress=True)
```

```python
dni_tygodnia_mapa = {
    'Monday': 1,
    'Tuesday': 2,
    'Wednesday': 3,
    'Thursday': 4,
    'Friday': 5,
    'Saturday': 6,
    'Sunday': 7
}
```

```python
dataset = pd.read_csv('inputData/ewh_profile[0].csv')

dataset
```

```python
dane = pd.DataFrame({"Pora_roku": [], "Dzien_tygodnia": [], "Czas_dnia": [], "Przeplyw":[]})
path = os.getcwd() 
csv_files = glob.glob(os.path.join(path, "inputData/wybrane/*.csv")) 
j=1;
for f in csv_files: 
    dt = pd.read_csv(f) 

    dane_1 = pd.DataFrame({"Pora_roku": [], "Dzien_tygodnia": [], "Czas_dnia": [], "Przeplyw":[]})
    dane_2 = pd.DataFrame({"Pora_roku": [], "Dzien_tygodnia": [], "Czas_dnia": [], "Przeplyw":[]})
    dane_3 = pd.DataFrame({"Pora_roku": [], "Dzien_tygodnia": [], "Czas_dnia": [], "Przeplyw":[]})
    dane_4 = pd.DataFrame({"Pora_roku": [], "Dzien_tygodnia": [], "Czas_dnia": [], "Przeplyw":[]})

    for i in range(len(dataset)):
        dt = dataset.loc[i,"Summer_Timestamps"]   
        data, time =dt.split(' ')
        year, month, day  = (int(x) for x in data.split('-'))    
        ans = datetime.date(year, month, day)

        hours, minutes,null= time.split(":")
        time=(int(hours)*60+int(minutes))/(60*24)

        dane_1.loc[len(dane_1)]={'Pora_roku':2,'Dzien_tygodnia':dni_tygodnia_mapa[ans.strftime("%A")],'Czas_dnia':time, 'Przeplyw': dataset.loc[i,"Summer_Water_Consumption"]}
    for i in range(len(dataset)):
        # ----------------------
        # ----------------------
        dt = dataset.loc[i,"Autumn_Timestamps"]   
        data, time =dt.split(' ')
        year, month, day  = (int(x) for x in data.split('-'))    
        ans = datetime.date(year, month, day)

        hours, minutes,null= time.split(":")
        time=(int(hours)*60+int(minutes))/(60*24)

        dane_2.loc[len(dane_2)]={'Pora_roku':3,'Dzien_tygodnia':dni_tygodnia_mapa[ans.strftime("%A")],'Czas_dnia':time, 'Przeplyw': dataset.loc[i,"Autumn_Water_Consumption"]}
    for i in range(len(dataset)):
        # #----------------------
        #----------------------
        dt = dataset.loc[i,"Winter_Timestamps"]   
        data, time =dt.split(' ')
        year, month, day  = (int(x) for x in data.split('-'))    
        ans = datetime.date(year, month, day)

        hours, minutes,null= time.split(":")
        time=(int(hours)*60+int(minutes))/(60*24)

        dane_3.loc[len(dane_3)]={'Pora_roku':4,'Dzien_tygodnia':dni_tygodnia_mapa[ans.strftime("%A")],'Czas_dnia':time, 'Przeplyw': dataset.loc[i,"Winter_Water_Consumption"]}
    for i in range(len(dataset)):
        # #----------------------
        # #----------------------
        dt = dataset.loc[i,"Spring_Timestamps"]   
        data, time =dt.split(' ')
        year, month, day  = (int(x) for x in data.split('-'))    
        ans = datetime.date(year, month, day)
    
        hours, minutes,null= time.split(":")
        time=(int(hours)*60+int(minutes))/(60*24)
    
        dane_4.loc[len(dane_4)]={'Pora_roku':1,'Dzien_tygodnia':dni_tygodnia_mapa[ans.strftime("%A")],'Czas_dnia':time, 'Przeplyw': dataset.loc[i,"Spring_Water_Consumption"]}
    dane= pd.concat([dane, dane_1,dane_2,dane_3,dane_4], ignore_index=True)
    dane.to_csv('dane_scalone__poj_backup.csv', index=False)
    print("DANE ITERACJA")
    print(j)
    j+=1
    print(dane)

print("KONIEC")
print(dane)
dane.to_csv('dane_poj_wybrane.csv', index=False)
```
