```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import LearningRateScheduler, TensorBoard
import matplotlib.pyplot as plt
import warnings

# Make NumPy printouts easier to read.
np.set_printoptions(precision=3, suppress=True)
warnings.filterwarnings("ignore")

print(tf.__version__)
```

```python
model_A_caly    = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Całość/model_a")
model_A_wybrany = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Wybrane/model_a")
model_A_poj     = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Pojedyńcze/model_a")

model_B_caly    = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Całość/model_b")
model_B_wybrany = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Wybrane/model_b")
model_B_poj     = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Pojedyńcze/model_b")

model_C_caly     = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Całość/model_c")
model_C_wybrany  = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Wybrane/model_c")
model_C_poj      = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Pojedyńcze/model_c")

model_D_caly    = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Całość/model_d")
model_D_wybrany = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Wybrane/model_d")
model_D_poj     = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Pojedyńcze/model_d")
```

## MODEL A

```python
flow_time=np.arange(0,100,100/1440) 
flow_time_plot=np.arange(0,24,24/1440) 

flow_time_pon=np.vstack((1*np.ones_like(flow_time, dtype=int),flow_time)).T
flow_result_pon_caly = model_A_caly.predict(flow_time_pon)  
flow_result_pon_wybrany = model_A_wybrany.predict(flow_time_pon)  
flow_result_pon_poj = model_A_poj.predict(flow_time_pon)  

flow_time_sr=np.vstack((3*np.ones_like(flow_time, dtype=int),flow_time)).T
flow_result_sr_caly = model_A_caly.predict(flow_time_sr)  
flow_result_sr_wybrany = model_A_wybrany.predict(flow_time_sr)  
flow_result_sr_poj = model_A_poj.predict(flow_time_sr)  

flow_time_sob=np.vstack((6*np.ones_like(flow_time, dtype=int),flow_time)).T
flow_result_sob_caly = model_A_caly.predict(flow_time_sob)  
flow_result_sob_wybrany = model_A_wybrany.predict(flow_time_sob)  
flow_result_sob_poj = model_A_poj.predict(flow_time_sob)

flow_time_nd=np.vstack((7*np.ones_like(flow_time, dtype=int),flow_time)).T
flow_result_nd_caly = model_A_caly.predict(flow_time_nd)  
flow_result_nd_wybrany = model_A_wybrany.predict(flow_time_nd)  
flow_result_nd_poj = model_A_poj.predict(flow_time_nd)

fig, axs = plt.subplots(3, 4, figsize=(16, 10))


axs[0,0].plot(flow_time_plot, flow_result_pon_caly)
axs[0,0].set_title("Dane: całość, Dzień tygodnia poniedziałek")
axs[0,1].plot(flow_time_plot, flow_result_sr_caly)
axs[0,1].set_title("Dane: całość, Dzień tygodnia środa")
axs[0,2].plot(flow_time_plot, flow_result_sob_caly)
axs[0,2].set_title("Dane: całość, Dzień tygodnia sobota")
axs[0,3].plot(flow_time_plot, flow_result_nd_caly)
axs[0,3].set_title("Dane: całość, Dzień tygodnia niedziela")

axs[1,0].plot(flow_time_plot, flow_result_pon_wybrany)
axs[1,0].set_title("Dane: wybrane, Dzień tygodnia poniedziałek")
axs[1,1].plot(flow_time_plot, flow_result_sr_wybrany)
axs[1,1].set_title("Dane: wybrane, Dzień tygodnia środa")
axs[1,2].plot(flow_time_plot, flow_result_sob_wybrany)
axs[1,2].set_title("Dane: wybrane, Dzień tygodnia sobota")
axs[1,3].plot(flow_time_plot, flow_result_nd_wybrany)
axs[1,3].set_title("Dane: wybrane, Dzień tygodnia niedziela")

axs[2,0].plot(flow_time_plot, flow_result_pon_poj)
axs[2,0].set_title("Dane: pojedyńcze, Dzień tygodnia poniedziałek")
axs[2,1].plot(flow_time_plot, flow_result_sr_poj)
axs[2,1].set_title("Dane: pojedyńcze, Dzień tygodnia środa")
axs[2,2].plot(flow_time_plot, flow_result_sob_poj)
axs[2,2].set_title("Dane: pojedyńcze, Dzień tygodnia sobota")
axs[2,3].plot(flow_time_plot, flow_result_nd_poj)
axs[2,3].set_title("Dane: pojedyńcze, Dzień tygodnia niedziela")


for i in range(3):
    for j in range(4):

        axs[i, j].set_xlabel('Czas, h')
        axs[i, j].set_ylabel('Przepływ, l/m')
        axs[i, j].set_xlim(0, 24)

plt.tight_layout()

plt.show()
```

## MODEL D

```python
flow_time=np.arange(0,100,100/1440) 
flow_time_plot=np.arange(0,24,24/1440) 

flow_result_spring=np.vstack((1*np.ones_like(flow_time, dtype=int),flow_time)).T
flow_result_spring_caly = model_D_caly.predict(flow_result_spring)  
flow_result_spring_wybrany = model_D_wybrany.predict(flow_result_spring)  
flow_result_spring_wybrany = np.round(flow_result_spring_wybrany, decimals=4) 
flow_result_spring_poj = model_D_poj.predict(flow_result_spring)  

flow_result_summer=np.vstack((2*np.ones_like(flow_time, dtype=int),flow_time)).T
flow_result_summer_caly = model_D_caly.predict(flow_result_spring)  
flow_result_summer_wybrany = model_D_wybrany.predict(flow_result_spring) 
flow_result_summer_wybrany = np.round(flow_result_summer_wybrany, decimals=4) 
flow_result_summer_poj = model_D_poj.predict(flow_result_spring) 

flow_result_fall=np.vstack((3*np.ones_like(flow_time, dtype=int),flow_time)).T
flow_result_fall_caly = model_D_caly.predict(flow_result_spring)  
flow_result_fall_wybrany = model_D_wybrany.predict(flow_result_spring)  
flow_result_fall_wybrany = np.round(flow_result_fall_wybrany, decimals=4) 
flow_result_fall_poj = model_D_poj.predict(flow_result_spring)   


fig, axs = plt.subplots(3, 3, figsize=(12, 8))
axs[0,0].plot(flow_time_plot, flow_result_summer_caly)
axs[0,0].set_title("Dane: całość, Pora roku: wiosna")
axs[0,1].plot(flow_time_plot, flow_result_spring_caly)
axs[0,1].set_title("Dane: całość, Pora roku: lato")
axs[0,2].plot(flow_time_plot, flow_result_fall_caly)
axs[0,2].set_title("Dane: całość, Pora roku: jesień")

axs[1,0].plot(flow_time_plot, flow_result_summer_wybrany)
axs[1,0].set_title("Dane: wybrane, Pora roku: wiosna")
axs[1,1].plot(flow_time_plot, flow_result_spring_wybrany)
axs[1,1].set_title("Dane: wybrane, Pora roku: lato")
axs[1,2].plot(flow_time_plot, flow_result_fall_wybrany)
axs[1,2].set_title("Dane: wybrane, Pora roku: jesień")

axs[2,0].plot(flow_time_plot, flow_result_summer_poj)
axs[2,0].set_title("Dane: pojedyńcze, Pora roku: wiosna")
axs[2,1].plot(flow_time_plot, flow_result_spring_poj)
axs[2,1].set_title("Dane: pojedyńcze, Pora roku: lato")
axs[2,2].plot(flow_time_plot, flow_result_fall_poj)
axs[2,2].set_title("Dane: pojedyńcze, Pora roku: jesień")

for i in range(3):
    for j in range(3):

        axs[i, j].set_xlabel('Czas, h', fontsize=11)
        axs[i, j].set_ylabel('Przepływ, l/m', fontsize=11)
        axs[i, j].set_xlim(0, 24)

plt.tight_layout()

plt.show()
```
