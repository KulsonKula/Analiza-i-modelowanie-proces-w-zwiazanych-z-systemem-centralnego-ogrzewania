```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import LearningRateScheduler, TensorBoard
import matplotlib.pyplot as plt
import warnings

# Make NumPy printouts easier to read.
np.set_printoptions(precision=3, suppress=True)
warnings.filterwarnings("ignore")

print(tf.__version__)
```

```python
model_A_caly    = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Całość/model_a")
model_A_wybrany = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Wybrane/model_a")
model_A_poj     = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Pojedyńcze/model_a")

model_B_caly    = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Całość/model_b")
model_B_wybrany = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Wybrane/model_b")
model_B_poj     = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Pojedyńcze/model_b")

model_C_caly     = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Całość/model_c")
model_C_wybrany  = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Wybrane/model_c")
model_C_poj      = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Pojedyńcze/model_c")

model_D_caly    = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Całość/model_d")
model_D_wybrany = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Wybrane/model_d")
model_D_poj     = tf.keras.models.load_model("../Modele/Tesowanie_architektury_sieci/Pojedyńcze/model_d")
```

## MODEL A

```python
flow_time=np.arange(0,100,100/1440) 
flow_time_plot=np.arange(0,24,24/1440) 

flow_time_pon=np.vstack((1*np.ones_like(flow_time, dtype=int),flow_time)).T
flow_result_pon_caly = model_A_caly.predict(flow_time_pon)  
flow_result_pon_wybrany = model_A_wybrany.predict(flow_time_pon)  
flow_result_pon_poj = model_A_poj.predict(flow_time_pon)  

flow_time_sr=np.vstack((3*np.ones_like(flow_time, dtype=int),flow_time)).T
flow_result_sr_caly = model_A_caly.predict(flow_time_sr)  
flow_result_sr_wybrany = model_A_wybrany.predict(flow_time_sr)  
flow_result_sr_poj = model_A_poj.predict(flow_time_sr)  

flow_time_sob=np.vstack((6*np.ones_like(flow_time, dtype=int),flow_time)).T
flow_result_sob_caly = model_A_caly.predict(flow_time_sob)  
flow_result_sob_wybrany = model_A_wybrany.predict(flow_time_sob)  
flow_result_sob_poj = model_A_poj.predict(flow_time_sob)

flow_time_nd=np.vstack((7*np.ones_like(flow_time, dtype=int),flow_time)).T
flow_result_nd_caly = model_A_caly.predict(flow_time_nd)  
flow_result_nd_wybrany = model_A_wybrany.predict(flow_time_nd)  
flow_result_nd_poj = model_A_poj.predict(flow_time_nd)

fig, axs = plt.subplots(4, 3, figsize=(16, 10))


axs[0,0].plot(flow_time_plot, flow_result_pon_caly,color='blue')
axs[0,0].set_title("Dane: całość, Dzień tygodnia poniedziałek")
axs[1,0].plot(flow_time_plot, flow_result_sr_caly,color='blue')
axs[1,0].set_title("Dane: całość, Dzień tygodnia środa")
axs[2,0].plot(flow_time_plot, flow_result_sob_caly,color='blue')
axs[2,0].set_title("Dane: całość, Dzień tygodnia sobota")
axs[3,0].plot(flow_time_plot, flow_result_nd_caly,color='blue')
axs[3,0].set_title("Dane: całość, Dzień tygodnia niedziela")

axs[0,1].plot(flow_time_plot, flow_result_pon_wybrany,color='blue')
axs[0,1].set_title("Dane: wybrane, Dzień tygodnia poniedziałek")
axs[1,1].plot(flow_time_plot, flow_result_sr_wybrany,color='blue')
axs[1,1].set_title("Dane: wybrane, Dzień tygodnia środa")
axs[2,1].plot(flow_time_plot, flow_result_sob_wybrany,color='blue')
axs[2,1].set_title("Dane: wybrane, Dzień tygodnia sobota")
axs[3,1].plot(flow_time_plot, flow_result_nd_wybrany,color='blue')
axs[3,1].set_title("Dane: wybrane, Dzień tygodnia niedziela")

axs[0,2].plot(flow_time_plot, flow_result_pon_poj,color='blue')
axs[0,2].set_title("Dane: pojedyńcze, Dzień tygodnia poniedziałek")
axs[1,2].plot(flow_time_plot, flow_result_sr_poj,color='blue')
axs[1,2].set_title("Dane: pojedyńcze, Dzień tygodnia środa")
axs[2,2].plot(flow_time_plot, flow_result_sob_poj,color='blue')
axs[2,2].set_title("Dane: pojedyńcze, Dzień tygodnia sobota")
axs[3,2].plot(flow_time_plot, flow_result_nd_poj,color='blue')
axs[3,2].set_title("Dane: pojedyńcze, Dzień tygodnia niedziela")


for i in range(4):
    for j in range(3):

        axs[i, j].set_xlabel('Czas, h', fontsize=11)
        axs[i, j].set_ylabel('Przepływ, l/m', fontsize=11)
        axs[i, j].set_xlim(0, 24)

plt.tight_layout()

plt.show()
```

```python
DATASET_wybrane=pd.read_csv('../Dane/data_nowe/dane_przetworzone/12_wybranych_nowy_parametr/calosc/calosc.csv')
DATASET_wybrane=DATASET_wybrane.drop("Tyg_zuzycie", axis=1)
DATASET_poj=pd.read_csv('../Dane/data_nowe/dane_przetworzone/12_wybranych_nowy_parametr/plik_1.csv')
DATASET_poj=DATASET_poj.drop("Tyg_zuzycie", axis=1)
DATASET_calosc=pd.read_csv('../../dane_scalone.csv')


y_calosc=DATASET_calosc.Przeplyw
y_wybrane=DATASET_wybrane.Przeplyw
y_poj=DATASET_poj.Przeplyw
```

```python
# modele A
a_wybrane=DATASET_wybrane.drop("Pora_roku", axis=1)
a_wybrane=a_wybrane.drop("Przeplyw", axis=1)

a_calosc=DATASET_calosc.drop("Pora_roku", axis=1)
a_calosc=a_calosc.drop("Przeplyw", axis=1)

a_poj=DATASET_poj.drop("Pora_roku", axis=1)
a_poj=a_poj.drop("Przeplyw", axis=1)

# modele b
b_wybrane=DATASET_wybrane.drop("Pora_roku", axis=1)
b_wybrane=b_wybrane.drop("Przeplyw", axis=1)
b_wybrane=b_wybrane.drop("Dzien_tygodnia", axis=1)

b_calosc=DATASET_calosc.drop("Pora_roku", axis=1)
b_calosc=b_calosc.drop("Przeplyw", axis=1)
b_calosc=b_calosc.drop("Dzien_tygodnia", axis=1)

b_poj=DATASET_poj.drop("Pora_roku", axis=1)
b_poj=b_poj.drop("Przeplyw", axis=1)
b_poj=b_poj.drop("Dzien_tygodnia", axis=1)

#modele c
c_wybrane=DATASET_wybrane.drop("Przeplyw", axis=1)

c_calosc=DATASET_calosc.drop("Przeplyw", axis=1)

c_poj=DATASET_poj.drop("Przeplyw", axis=1)

#modele d

d_wybrane=DATASET_wybrane.drop("Przeplyw", axis=1)
d_wybrane=d_wybrane.drop("Dzien_tygodnia", axis=1)

d_calosc=DATASET_calosc.drop("Przeplyw", axis=1)
d_calosc=d_calosc.drop("Dzien_tygodnia", axis=1)

d_poj=DATASET_poj.drop("Przeplyw", axis=1)
d_poj=d_poj.drop("Dzien_tygodnia", axis=1)



x_train_calosc_A, x_test_calosc_A, y_train_calosc_A, y_test_calosc_A = train_test_split(a_calosc, y_calosc, test_size=0.2, random_state=42)
x_train_wybrane_A, x_test_wybrane_A, y_train_wybrane_A, y_test_wybrane_A = train_test_split(a_wybrane, y_wybrane, test_size=0.2, random_state=42)
x_train_poj_A, x_test_poj_A, y_train_poj_A, y_test_poj_A = train_test_split(a_poj, y_poj, test_size=0.2, random_state=42)

x_train_calosc_B, x_test_calosc_B, y_train_calosc_B, y_test_calosc_B = train_test_split(b_calosc, y_calosc, test_size=0.2, random_state=42)
x_train_wybrane_B, x_test_wybrane_B, y_train_wybrane_B, y_test_wybrane_B = train_test_split(b_wybrane, y_wybrane, test_size=0.2, random_state=42)
x_train_poj_B, x_test_poj_B, y_train_poj_B, y_test_poj_B = train_test_split(b_poj, y_poj, test_size=0.2, random_state=42)

x_train_calosc_C, x_test_calosc_C, y_train_calosc_C, y_test_calosc_C = train_test_split(c_calosc, y_calosc, test_size=0.2, random_state=42)
x_train_wybrane_C, x_test_wybrane_C, y_train_wybrane_C, y_test_wybrane_C = train_test_split(c_wybrane, y_wybrane, test_size=0.2, random_state=42)
x_train_poj_C, x_test_poj_C, y_train_poj_C, y_test_poj_C = train_test_split(c_poj, y_poj, test_size=0.2, random_state=42)

x_train_calosc_D, x_test_calosc_D, y_train_calosc_D, y_test_calosc_D = train_test_split(d_calosc, y_calosc, test_size=0.2, random_state=42)
x_train_wybrane_D, x_test_wybrane_D, y_train_wybrane_D, y_test_wybrane_D = train_test_split(d_wybrane, y_wybrane, test_size=0.2, random_state=42)
x_train_poj_D, x_test_poj_D, y_train_poj_D, y_test_poj_D = train_test_split(d_poj, y_poj, test_size=0.2, random_state=42)
```

```python
BATCH_SIZE=64
model_A_caly.evaluate(x_test_calosc_A, y_test_calosc_A, batch_size=BATCH_SIZE)
model_A_wybrany.evaluate(x_test_wybrane_A, y_test_wybrane_A, batch_size=BATCH_SIZE)
model_A_poj.evaluate(x_test_poj_A, y_test_poj_A, batch_size=BATCH_SIZE)

model_B_caly.evaluate(x_test_calosc_B, y_test_calosc_B, batch_size=BATCH_SIZE)
model_B_wybrany.evaluate(x_test_wybrane_B, y_test_wybrane_B, batch_size=BATCH_SIZE)
model_B_poj.evaluate(x_test_poj_B, y_test_poj_B, batch_size=BATCH_SIZE)

model_C_caly.evaluate(x_test_calosc_C, y_test_calosc_C, batch_size=BATCH_SIZE)
model_C_wybrany.evaluate(x_test_wybrane_C, y_test_wybrane_C, batch_size=BATCH_SIZE)
model_C_poj.evaluate(x_test_poj_C, y_test_poj_C, batch_size=BATCH_SIZE)

model_D_caly.evaluate(x_test_calosc_D, y_test_calosc_D, batch_size=BATCH_SIZE)
model_D_wybrany.evaluate(x_test_wybrane_D, y_test_wybrane_D, batch_size=BATCH_SIZE)
model_D_poj.evaluate(x_test_poj_D, y_test_poj_D, batch_size=BATCH_SIZE)
```
