```python
import tensorflow as tf
import pandas as pd
import numpy as np
import datetime
import os 
import glob 
np.set_printoptions(precision=5, suppress=True)
```

```python
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
dni_tygodnia_mapa = {
    'Monday': 1,
    'Tuesday': 2,
    'Wednesday': 3,
    'Thursday': 4,
    'Friday': 5,
    'Saturday': 6,
    'Sunday': 7
}
```

```python
j=1;
n=1440*7
dane = pd.DataFrame({"Pora_roku": [], "Dzien_tygodnia": [], "Czas_dnia": [], "Przeplyw":[], "Tyg_zuzycie":[]})
calosc=pd.DataFrame({"Pora_roku": [], "Dzien_tygodnia": [], "Czas_dnia": [], "Przeplyw":[], "Tyg_zuzycie":[]})
path = os.getcwd() 
csv_files = glob.glob(os.path.join(path, "inputData/wybrane/*.csv")) 

for f in csv_files: 
    dataset = pd.read_csv(f) 

    dane_1 = pd.DataFrame({"Pora_roku": [], "Dzien_tygodnia": [], "Czas_dnia": [], "Przeplyw":[], "Tyg_zuzycie":[]})
    dane_2 = pd.DataFrame({"Pora_roku": [], "Dzien_tygodnia": [], "Czas_dnia": [], "Przeplyw":[], "Tyg_zuzycie":[]})
    dane_3 = pd.DataFrame({"Pora_roku": [], "Dzien_tygodnia": [], "Czas_dnia": [], "Przeplyw":[], "Tyg_zuzycie":[]})
    dane_4 = pd.DataFrame({"Pora_roku": [], "Dzien_tygodnia": [], "Czas_dnia": [], "Przeplyw":[], "Tyg_zuzycie":[]})

    k=0
    for i in range(len(dataset)):
        dt = dataset.loc[i,"Summer_Timestamps"]   
        data, time =dt.split(' ')
        year, month, day  = (int(x) for x in data.split('-'))    
        ans = datetime.date(year, month, day)

        hours, minutes,null= time.split(":")
        time=(int(hours)*60+int(minutes))/(60*24)

        water_usage=dataset["Summer_Water_Consumption"].loc[n*k:n*(k+1)].sum()
        dane_1.loc[len(dane_1)]={'Pora_roku':2,'Dzien_tygodnia':dni_tygodnia_mapa[ans.strftime("%A")],'Czas_dnia':time, 'Przeplyw': dataset.loc[i,"Summer_Water_Consumption"],'Tyg_zuzycie':water_usage}
        if(i%n==0 and i!=0):
            k+=1

    k=0
    for i in range(len(dataset)):
        # ----------------------
        # ----------------------
        dt = dataset.loc[i,"Autumn_Timestamps"]   
        data, time =dt.split(' ')
        year, month, day  = (int(x) for x in data.split('-'))    
        ans = datetime.date(year, month, day)

        hours, minutes,null= time.split(":")
        time=(int(hours)*60+int(minutes))/(60*24)

        water_usage=dataset["Autumn_Water_Consumption"].loc[n*k:n*(k+1)].sum()
        dane_2.loc[len(dane_2)]={'Pora_roku':3,'Dzien_tygodnia':dni_tygodnia_mapa[ans.strftime("%A")],'Czas_dnia':time, 'Przeplyw': dataset.loc[i,"Autumn_Water_Consumption"],'Tyg_zuzycie':water_usage}
        if(i%n==0 and i!=0):
            k+=1

    k=0
    for i in range(len(dataset)):
        # #----------------------
        #----------------------
        dt = dataset.loc[i,"Winter_Timestamps"]   
        data, time =dt.split(' ')
        year, month, day  = (int(x) for x in data.split('-'))    
        ans = datetime.date(year, month, day)

        hours, minutes,null= time.split(":")
        time=(int(hours)*60+int(minutes))/(60*24)

        water_usage=dataset["Winter_Water_Consumption"].loc[n*k:n*(k+1)].sum()
        dane_3.loc[len(dane_3)]={'Pora_roku':4,'Dzien_tygodnia':dni_tygodnia_mapa[ans.strftime("%A")],'Czas_dnia':time, 'Przeplyw': dataset.loc[i,"Winter_Water_Consumption"],'Tyg_zuzycie':water_usage}
        if(i%n==0 and i!=0):
            k+=1

    k=0
    for i in range(len(dataset)):
        # #----------------------
        # #----------------------
        dt = dataset.loc[i,"Spring_Timestamps"]   
        data, time =dt.split(' ')
        year, month, day  = (int(x) for x in data.split('-'))    
        ans = datetime.date(year, month, day)
    
        hours, minutes,null= time.split(":")
        time=(int(hours)*60+int(minutes))/(60*24)
    
        water_usage=dataset["Spring_Water_Consumption"].loc[n*k:n*(k+1)].sum()
        dane_4.loc[len(dane_4)]={'Pora_roku':1,'Dzien_tygodnia':dni_tygodnia_mapa[ans.strftime("%A")],'Czas_dnia':time, 'Przeplyw': dataset.loc[i,"Spring_Water_Consumption"],'Tyg_zuzycie': water_usage}
        if(i%n==0 and i!=0):
            k+=1

    dane= pd.concat([dane_1,dane_2,dane_3,dane_4], ignore_index=True)
    dane.to_csv('dane_przetworzone/12_wybranych/plik_'+str(j)+'.csv', index=False)
    calosc=pd.concat([calosc, dane_1,dane_2,dane_3,dane_4], ignore_index=True)
    print("DANE ITERACJA")
    print(j)
    print(dane)
    j+=1

print("KONIEC")
print(dane)
calosc.to_csv('dane_przetworzone/12_wybranych/calosc.csv', index=False)
```
